```python
a_file = open('data.txt','r')
data = a_file.read()
a_file.close()

import matplotlib.pyplot as plt

data = data.replace('.', '')
data = data.replace(',', '')
data = data.replace('-', '')
data = data.lower()
data = data.split()
dictionay = {}
for word in data:
    dictionay[word] = dictionay.get(word,0)+1
names = list(dictionay.keys())
values = list(dictionay.values())

plt.bar(range(5), values[:5], tick_label = names[:5])

```

```python
import cv2
import numpy as np
img_cat = cv2.imread('binary_cat.png',0)
print(np.unique(img_cat))

counts = {}
w,h = img_cat.shape
for row in range(w):
    for col in range(h):
        counts[img_cat[row,col]] = counts.get(img_cat[row,col],0)+1
values = list(counts.values())
names = list(counts.keys())
plt.bar(range(len(counts)), values, tick_label = names)
```
